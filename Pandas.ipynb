# Задание 2 #
**По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.**

Фильтр получился очень массивным


```python
energy = pd.read_csv('/Users/klavdiakavesnikova/projects/практики к лекциям/10._Основы_pandas/power.csv')
```

```python
filtered_countries = energy[ ((energy['country']=='Lithuania') | (energy['country']=='Latvia') | (energy['country']=='Estonia')) & (energy['category'].isin([4, 12, 21])) & ((energy['year'] < 2011) | (energy['year'] > 2005))]
filtered_countries.head()
```

```python
filtered_countries_positive_quantity = filtered_countries[ (filtered_countries['quantity'] >= 0)]
filtered_countries_positive_quantity['quantity'].sum()
```
